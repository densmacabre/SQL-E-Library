```python
# Import Modules
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Create soup object
url = "https://books.toscrape.com/index.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Get information about books genre
# Determine and assign html tag
genre_tag = soup.find('ul', class_='nav nav-list')
genres = genre_tag.find_all('ul')

# Store list of book genre
book_genre = []

# Loop elements in ul tag
for genre in genres:
    genre_list = genre.find_all('li')

    for name in genre_list:
        genre_name = name.find('a')

        if genre_name is not None:
            genre_of_books = genre_name.text.strip()
            book_genre.append(genre_of_books)

# Answering question
print(f"There are {len(book_genre)} genres in this store, there are as follows:")
for books in book_genre:
    print(books)
```

```python
# It seems that there are books of unspecified genre
# Determine unspecified genre
print(f"It seems that book from {book_genre[13]} and {book_genre[16]} genre are not specified")
```

```python
# How many books are in unspecified genre pt.1
# Books from Default genre
page_counter_1 = 1 # Page counter
book_titles_1 = []

# Loop every page based on counter above
while True:
    url_genre_1 = f"https://books.toscrape.com/catalogue/category/books/default_15/page-{page_counter_1}.html"
    response_1 = requests.get(url_genre_1)
    soup_1 = BeautifulSoup(response_1.content, 'html.parser')
    
    # Check if there is no more page for handling None
    ol_1 = soup_1.find('ol')
    if ol_1 is None:
        break
    
    # Find all article tags inside the ol tag
    article_1 = ol_1.find_all('article', class_='product_pod')

    # Loop article tags to find titles
    for a1 in article_1:
        image_1 = a1.find('img')
        title_1 = image_1['alt']
        book_titles_1.append(title_1)
    
    page_counter_1 += 1 # Add number to counter

# Answering question
print(f"There are {len(book_titles_1)} books from this genre")
```

```python
# How many books are in unspecified genre pt.2
# Books from Add a comment genre
page_counter_2 = 1 # Page counter
book_titles_2 = []

# Loop every page based on counter above
while True:
    url_genre_2 = f"https://books.toscrape.com/catalogue/category/books/add-a-comment_18/page-{page_counter_2}.html"
    response_2 = requests.get(url_genre_2)
    soup_2 = BeautifulSoup(response_2.content, 'html.parser')
    
    # Check if there is no more page for handling None
    ol_2 = soup_2.find('ol')
    if ol_2 is None:
        break
    
    # Find all article tags inside the ol tag
    article_2 = ol_2.find_all('article', class_='product_pod')

    # Loop article tags to find titles
    for a2 in article_2:
        image_2 = a2.find('img')
        title_2 = image_2['alt']
        book_titles_2.append(title_2)
    
    page_counter_2 += 1 # Add number to counter

# Answering question
print(f"There are {len(book_titles_2)} books from this genre")
```

```python
# Get URLs for every genre of books to loop inside a list
# Base URL
base_url = "https://books.toscrape.com/"

# Main URL
main_url = base_url + "catalogue/category/books_1/index.html"

# Create soup object
response_3 = requests.get(main_url)
soup_3 = BeautifulSoup(response_3.text, 'html.parser')

# Search for links tag
genre_tag = soup_3.find('ul', class_='nav nav-list')
links_tag = genre_tag.find_all('a')

# Store list of book genre
genre_links = []

# Loop elements in a tag
for link in links_tag:
    href_name = link.text.strip()
    href_links = link['href']
    href_links = href_links[2:]
    
    full_genre_links = base_url + "catalogue/category" + href_links
    genre_links.append(full_genre_links)

genre_links = genre_links[1:]
genre_links
```

```python
# Exclude links of unspecified genre
# Create a dictionary to map links into its genre
genre_to_dict = book_genre
links_to_dict = genre_links
links_dict = {k: v for (k, v) in zip(genre_to_dict, links_to_dict)}
links_dict
```

```python
# Exclude links of unspecified genre
to_exclude = ['Default', 'Add a comment']
links_to_use = {k: v for (k, v) in links_dict.items() if k not in to_exclude}
links_to_use
```

```python
# Extract books title and its genre from URLs above
# Create empty list to store extracted data
all_books_data = []

# Loop URLs in links_to_use dict
for genre, link in links_to_use.items():
    page_number = 1

    while True:
        
        if page_number == 1:
            current_url = link
        else:
            current_url = link.replace('index.html', f'page-{page_number}.html')
        
        current_response = requests.get(current_url)

        if current_response.status_code == 400:
            print(f"Reached end of pages for genre: {genre}")
            break

        current_soup = BeautifulSoup(current_response.text, 'html.parser')

        # Find the ordered list tag
        current_ol = current_soup.find('ol')

        # Break the loop if there is no ol tag
        if not current_ol:
            break

        # Extract books
        for article in current_ol.find_all('article', class_='product_pod'):
            title = article.find('img')['alt']
            all_books_data.append({'title': title, 'genre': genre})  # Only title and genre
        
        page_number += 1 # Add this to page number counter

# Create the final DataFrame
books_df = pd.DataFrame(all_books_data, columns=['title', 'genre'])
print(books_df.head())
```

```python
# Save DataFrame to csv
books_df.to_csv('books_data.csv', index=False)
```
